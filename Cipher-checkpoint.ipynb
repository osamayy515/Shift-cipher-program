```python
from string  import ascii_uppercase

def get_char(s):
    LETTERS = {index: str(letter) for index, letter in enumerate(ascii_uppercase, start=0)} 
    return LETTERS[s]

def get_num(s):
    
    LETTERS = {str(letter): index for index, letter in enumerate(ascii_uppercase, start=0)} 
    return LETTERS[s.upper()]
    
```

```python
def decrypt(string,key):
    text=''
    string = string.upper()
    for s in string:
        if s==' ':
            text += s
        else:
            R = ((ord(s)-65)-key)%26
            
            if R<0:
                R=26-abs(R)
                R=R%26
            
            text += get_char(R)
        
    return text
```

# Affine Cipher

```python
C = (Pi . a) + b mod m
P = a-1 (Ci - b) mod m


Modulo Inverse
A . A-1 = 1 mod m
```

```python
def af_encrypt(string,a,b):
    cipher=''
    
    for s in string:
        if s==' ':
            cipher += s
        else:
            
            C=((get_num(s)*a)+b)%26
            if C<0:
                C=26-abs(C)
                C=C%26
            cipher += get_char(C)
            
    return cipher
```

```python
af_encrypt('a b',7,10)
```

```python
import math
def af_decrypt(string,a,b):
    text=''
    string=string.upper()
    
    for i in range(1,200):
        a_inv=(a*i)%26
        if a_inv==1:
            break
            
    for s in string:
        if s==' ':
            text += s
        else:
            R=(a_inv*(get_num(s)-b))%26
            if R<0:
                R=26-abs(R)
                R=R%26
                    
            text += get_char(R)
                
    return text
```

```python
af_encrypt('a',7,10)
```

```python
af_decrypt('k',7,10)
```
